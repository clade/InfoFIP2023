```python
from matplotlib.pyplot import figure
import numpy as np
from scipy.optimize import curve_fit
```

```python
frequence, probabilite = np.loadtxt('fit_sinus.dat', unpack=True)
```

```python
fig = figure()
ax = fig.subplots(1, 1)

ax.plot(frequence, probabilite, 'o')
```

```python
def fit_function(freq, moyenne, amplitude, centre, interfrange):
    return moyenne + amplitude*np.cos(2*np.pi*(freq - centre)/interfrange)

x_plot = np.linspace(-220, 220)

fig = figure()
ax = fig.subplots(1, 1)

p0 = [0.2, 0.1, 0, 200]

ax.plot(frequence, probabilite, 'o', label='experiment')
#ax.plot(x_plot, fit_function(x_plot, *p0))

popt, pcov = curve_fit(fit_function, frequence, probabilite, p0=p0)

ax.plot(x_plot, fit_function(x_plot, *popt), label='fit')

ax.set_xlabel('Frequence [Hz]')
ax.set_ylabel('Probabilite')
ax.legend()
```
