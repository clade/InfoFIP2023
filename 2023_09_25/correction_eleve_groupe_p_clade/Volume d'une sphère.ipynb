```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 1000
x = np.random.rand(M)*2 - 1
y = np.random.rand(M)*2 - 1

def distance(x, y):
    return np.sqrt(x**2 + y**2)

x_int=x[distance(x, y)<1]
y_int=y[distance(x, y)<1]

x_ext=x[distance(x, y)>1]
y_ext=y[distance(x, y)>1]

plt.scatter(x_int, y_int)
plt.scatter(x_ext, y_ext)
ax = plt.gca()
ax.set_aspect('equal')
```

```python
def volume(dimension, M):
    pts = np.random.rand(M, dimension)*2 - 1
    carre_norme = np.sum(pts**2, axis=1)
    return np.mean(carre_norme<1) * 2**dimension

volume(5, 1000000)
```

```python
%timeit volume(5, 10000000)
```

```python
data = np.array([volume(2, 10000) for _ in range(100)])
```
