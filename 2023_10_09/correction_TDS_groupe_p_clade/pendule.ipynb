```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def f(t, y):
    theta, theta_point = y
    return (theta_point, -np.sin(theta))
```

```python
y_0  = [np.pi/2, 0]
t_f = 10
res = solve_ivp(f, [0, t_f], y_0, t_eval=np.linspace(0, t_f), method='RK45', 
                rtol=1E-10, atol=1E-10)
```

```python
plt.plot(res.t, res.y[0])
```

```python
def energie(theta, theta_point):
    return theta_point**2/2 - np.cos(theta)

theta, theta_point = res.y

plt.plot(res.t, energie(theta, theta_point))
```
